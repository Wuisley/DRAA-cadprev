```python
import pandas as pd
import pyautogui
import pyperclip
import time
```

```python
df = pd.read_excel('plano de amortização.xlsx', sheet_name = 5)
df
```

```python
#TRATANDO DADOS

#FILTRANDO O GRUPO CIVIL E MILITAR
df = df.loc[df['Carreira']== "MILITAR"]

#REMOVENDO AS LINHAS COM DADOS EM BRANCO

df
```

```python
time.sleep(5)
pyautogui.press('tab')
#ESTATISTICA DA POPULAÇÃO COBERTA


for i in df["cod_orgao"].index:
    pyautogui.PAUSE = 0.4
    cod_org = df.loc[i, "cod_orgao"]
    tipo = df.loc[i,'cod_tipo']
    carreira = df.loc[i, 'cod_carreira']
    criterio = df.loc[i, 'cod_criterio']
    
    quantidade_m = df.loc[i, 'Qtd_m']
    media_beneficio_m = df.loc[i, 'Salário médio_m']
    im_m = df.loc[i, 'IM_m']
    imp_m = df.loc[i, 'IM Ap_m']
    ima_m = df.loc[i, 'IM Ad_m']
    
    quantidade_f = df.loc[i, 'Qtd_f']
    media_beneficio_f = df.loc[i, 'Salário médio_f']
    im_f = df.loc[i, 'IM_f']
    imp_f = df.loc[i, 'IM Ap_f']
    ima_f = df.loc[i, 'IM Ad_f']

    for n in range(cod_org):
        pyautogui.press('down')
    pyautogui.press('tab')  

    for n in range(int(carreira)):
        pyautogui.press('down')
    pyautogui.press('tab')

    for n in range(tipo):
        pyautogui.press('down')
    pyautogui.press('tab')

    for n in range(int(criterio)):
        pyautogui.press('down')
    pyautogui.press('tab')

# MASCULINO
    pyperclip.copy(int(quantidade_m))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyperclip.copy(str(media_beneficio_m))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyperclip.copy(str(im_m))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyperclip.copy(str(imp_m))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyperclip.copy(str(ima_m))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

# FEMININO
    pyperclip.copy(int(quantidade_f))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyperclip.copy(str(media_beneficio_f))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyperclip.copy(str(im_f))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyperclip.copy(str(imp_f))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyperclip.copy(str(ima_f))
    pyautogui.hotkey('ctrl', 'v')
    pyautogui.press('tab')

    pyautogui.press('space')

# VOLTANDO AO TOPO
    
    for t in range(12):
        pyautogui.PAUSE = 0.3
        pyautogui.hotkey('shift', 'tab')
```
