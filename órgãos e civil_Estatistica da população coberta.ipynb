```python
import pandas as pd
import pyautogui
import pyperclip
import time
```

```python
#IMPORTANDO BANCO DE DADOS

df = pd.read_excel('plano de amortização.xlsx', sheet_name = 4)
df2 = pd.read_excel('plano de amortização.xlsx', sheet_name = 5)
```

```python
#PREENCHENDO OS ÓRGÃOS/ENTIDADES

time.sleep(4)
pyautogui.PAUSE = 0.1
for v in range(4):
    pyautogui.press('tab')

pyautogui.press('space')

for b in range(4):
    pyautogui.hotkey('shift', 'tab')

for i, dados in enumerate(df["CNPJ"]):
    cnpj = df.loc[i, "CNPJ"]
    nome = df.loc[i, "NOME"]
    competencia = df.loc[i, "COMPETENCIA"]
    poder = df.loc[i, "PODER"]
    tipo = df.loc[i, "TIPO"]

    pyperclip.copy(str(cnpj))
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('Tab')

    pyperclip.copy(str(nome))
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('Tab')

    pyperclip.copy(str(competencia))
    pyautogui.hotkey('ctrl','v')
    pyautogui.press('Tab')

    pyautogui.hotkey('alt', 'down')
    for n in range(poder):
        pyautogui.press('down')

    pyautogui.press('Tab')
      
    for t in range(tipo-1):
        pyautogui.press('down')
    
    pyautogui.press('Tab')
    pyautogui.press('enter')   

    pyautogui.hotkey('shift','tab')

    for t in range(tipo-1):
        pyautogui.press('up')
    pyautogui.hotkey('shift','tab')
    pyautogui.hotkey('shift','tab')
    pyautogui.hotkey('shift','tab')
    pyautogui.hotkey('shift','tab')
```
